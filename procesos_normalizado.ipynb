```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
from datetime import datetime
from matplotlib.dates import DateFormatter
import matplotlib.ticker as ticker

# Configuración de estilo para las gráficas
plt.style.use('ggplot')
sns.set(font_scale=1.2)
sns.set_style("whitegrid")

# Configuración del sistema (AGREGAR ESTA VARIABLE)
LOGICAL_PROCESSORS = 16  # Tu sistema tiene 16 procesadores lógicos

# Definir las rutas de los archivos CSV (deberás ajustar estas rutas a tu estructura de archivos)
# Esta es una estructura de ejemplo, ajusta según tu organización de archivos
data_files = {
    'JADE': {
        'Small': 'Perfmon\JADE-SMALL-NEW\Procesos_194925024927.csv',  # PID: 92572
        'Medium': 'Perfmon\JADE-MEDIUM-NEW\Procesos_195025025003.csv',  # PID: 93596
        'Full': 'Perfmon\JADE-FULL-NEW\Procesos_195025025038.csv'      # PID: 66068
    },
    'SPADE': {
        'Small': 'Perfmon\SPADE-SMALL-NEW\Procesos_194125024111.csv',  # PID: 96052
        'Medium': 'Perfmon\SPADE-MEDIUM-NEW\Procesos_194425024408.csv', # PID: 77760
        'Full': 'Perfmon\SPADE-FULL-NEW\Procesos_194425024458.csv'     # PID: 32448
    }
}

# PIDs para cada plataforma y escenario
pids = {
    'JADE': {
        'Small': 92572,
        'Medium': 93596,
        'Full': 66068
    },
    'SPADE': {
        'Small': 96052,
        'Medium': 77760,
        'Full': 32448
    }
}

# Función para cargar y preprocesar datos de Perfmon (MODIFICADA CON CPU REAL)
def load_perfmon_data(filepath):
    """Carga y preprocesa un archivo CSV de Perfmon con cálculo de CPU real"""
    print(f"Cargando datos desde {filepath}...")
    
    try:
        # Leer el CSV
        df = pd.read_csv(filepath)
        
        # Renombrar la primera columna a 'Timestamp'
        if len(df.columns) > 0:
            df = df.rename(columns={df.columns[0]: 'Timestamp'})
        else:
            print(f"Error: El archivo {filepath} no tiene columnas.")
            return None
        
        for col in df.columns:
            if col != 'Timestamp':  # No intentar convertir la columna de timestamp
                try:
                    # Reemplazar comas y eliminar unidades si es necesario
                    df[col] = df[col].astype(str)
                    df[col] = df[col].str.replace(',', '')
                    df[col] = df[col].str.replace(' KB', '')
                    df[col] = df[col].str.replace(' MB', '')
                    df[col] = df[col].str.replace(' %', '')
                    
                    # Convertir a numérico, con errores como NaN
                    df[col] = pd.to_numeric(df[col], errors='coerce')
                except Exception as e:
                    print(f"Advertencia: No se pudo convertir la columna {col} a numérico: {e}")
        
        # 🔧 NUEVA FUNCIONALIDAD: Calcular CPU REAL (estilo Task Manager)
        cpu_columns = [col for col in df.columns if '% Processor Time' in col or 'tiempo de procesador' in col.lower()]
        total_cpu_cols = [col for col in df.columns if '_Total' in col and ('% Processor Time' in col or 'tiempo de procesador' in col.lower())]
        
        if total_cpu_cols:
            total_cpu_col = total_cpu_cols[0]
            print(f"✅ Encontrada columna _Total: {total_cpu_col}")
            
            for col in cpu_columns:
                if '_Total' not in col:  # No procesar la columna _Total sobre sí misma
                    # ✅ FÓRMULA CORRECTA: (proceso_cpu / (núcleos × total_cpu)) × 100
                    try:
                        real_cpu = (df[col] / (LOGICAL_PROCESSORS * df[total_cpu_col])) * 100
                        df[col + '_Real'] = real_cpu
                        print(f"✅ CPU REAL creada: {col}_Real")
                        print(f"   Máximo proceso: {df[col].max():.1f}%")
                        print(f"   Máximo real: {real_cpu.max():.2f}%")
                        print(f"   Promedio real: {real_cpu.mean():.2f}%")
                    except Exception as e:
                        print(f"❌ Error calculando CPU real para {col}: {e}")
            
            # También crear versión "normalizada simple" para comparación
            for col in cpu_columns:
                if '_Total' not in col:
                    df[col + '_Normalized_Simple'] = df[col] / LOGICAL_PROCESSORS
        else:
            print("⚠️  ADVERTENCIA: No se encontró columna _Total para CPU del sistema")
            print("   Solo se creará normalización simple (incorrecta)")
            # Fallback a normalización simple
            for col in cpu_columns:
                df[col + '_Normalized_Simple'] = df[col] / LOGICAL_PROCESSORS
        
        # Intentar extraer timestamps de la primera columna
        try:
            # Buscar patrones como MM/DD/YYYY HH:MM:SS.mmm en los datos
            date_pattern = r'(\d{1,4}[-/]\d{1,2}[-/]\d{1,4}\s\d{1,2}:\d{2}:\d{2}(?:\.\d+)?)'
            
            # Extraer timestamps si coinciden con el patrón
            timestamps = []
            for time_str in df['Timestamp']:
                match = re.search(date_pattern, str(time_str))
                if match:
                    try:
                        timestamp = pd.to_datetime(match.group(1))
                        timestamps.append(timestamp)
                    except:
                        timestamps.append(pd.NaT)
                else:
                    timestamps.append(pd.NaT)
                    
            # Si tenemos timestamps válidos, usarlos
            if not all(pd.isna(timestamps)):
                df['DateTimeIndex'] = timestamps
                df = df[~pd.isna(df['DateTimeIndex'])]
                print(f"Se extrajeron timestamps con éxito.")
            else:
                print("No se pudieron extraer timestamps de la primera columna. Usando secuencia numérica.")
                df['DateTimeIndex'] = pd.date_range(start=datetime.now(), periods=len(df), freq='S')
        except Exception as e:
            print(f"Error al procesar timestamps: {e}")
            print("Usando secuencia numérica con timestamps artificiales.")
            df['DateTimeIndex'] = pd.date_range(start=datetime.now(), periods=len(df), freq='S')
        
        # Calcular tiempo transcurrido en segundos desde el primer timestamp
        if 'DateTimeIndex' in df.columns:
            start_time = df['DateTimeIndex'].min()
            df['ElapsedTime'] = (df['DateTimeIndex'] - start_time).dt.total_seconds()
            print("Agregada columna 'ElapsedTime' en segundos.")
        else:
            print("Warning: No se pudo crear la columna 'DateTimeIndex'. Creando 'ElapsedTime' como secuencia simple.")
            df['ElapsedTime'] = range(len(df))
        
        return df
    
    except Exception as e:
        print(f"Error al cargar el archivo {filepath}: {e}")
        return None

# Función para encontrar columnas por PID o nombre de proceso
def find_columns_by_pid(df, pid, metric=None):
    """Encuentra columnas relacionadas con un PID específico y opcionalmente una métrica específica"""
    pid_cols = []
    
    if df is None:
        print(f"Warning: DataFrame es None al buscar columnas con PID {pid}")
        return pid_cols
    
    # Buscar columnas que contengan el PID
    pid_str = str(pid)
    for col in df.columns:
        if pid_str in col:
            if metric is None or metric in col:
                pid_cols.append(col)
    
    # Si no se encontraron columnas con el PID, buscar por nombre de proceso (java para JADE, python para SPADE)
    if not pid_cols:
        # Para JADE (Java)
        java_cols = []
        for col in df.columns:
            if 'java' in col.lower():
                if metric is None or metric in col:
                    java_cols.append(col)
        
        # Para SPADE (Python)
        python_cols = []
        for col in df.columns:
            if 'python' in col.lower():
                if metric is None or metric in col:
                    python_cols.append(col)
        
        # Determinar qué conjunto usar basado en el PID (asumiendo que PIDs de JADE y SPADE corresponden a procesos específicos)
        is_jade_pid = any(pid == p for p in pids['JADE'].values())
        if is_jade_pid and java_cols:
            pid_cols = java_cols
        elif not is_jade_pid and python_cols:
            pid_cols = python_cols
    
    if not pid_cols:
        print(f"Warning: No se encontraron columnas para el PID {pid}")
        
    return pid_cols

# 🔧 FUNCIÓN NUEVA: Buscar columnas de CPU REAL
def find_real_cpu_columns(df, pid):
    """Encuentra columnas de CPU real (_Real) para un PID específico"""
    pid_cols = find_columns_by_pid(df, pid, '% Processor Time')
    real_cols = []
    
    for col in pid_cols:
        real_col = col + '_Real'
        if real_col in df.columns:
            real_cols.append(real_col)
    
    return real_cols

# Función para limpiar nombres de columna para visualización
def clean_column_name(col_name, platform, scenario):
    """Extrae el nombre de proceso y métrica de un nombre de columna para visualización más limpia"""
    try:
        # Patrón para encontrar proceso(nombre)\métrica o formatos similares
        match = re.search(r'Process\((.*?)\)\\(.*)', col_name)
        if match:
            process_name = match.group(1)
            metric = match.group(2)
            return f"{platform} ({scenario}) - {process_name} - {metric}"
        else:
            parts = col_name.split('\\')
            if len(parts) > 1:
                return f"{platform} ({scenario}) - {parts[-1]}"
            else:
                return f"{platform} ({scenario}) - {col_name}"
    except:
        return f"{platform} ({scenario}) - {col_name}"

# Función para crear gráficas comparativas de CPU (ORIGINAL - sin cambios)
def plot_cpu_comparison(dataframes, platform_colors=None):
    """
    Crea una gráfica comparativa de uso de CPU entre JADE y SPADE para los tres escenarios
    
    Args:
        dataframes: diccionario de DataFrames organizados por plataforma y escenario
        platform_colors: diccionario opcional de colores para cada plataforma
    """
    if platform_colors is None:
        platform_colors = {'JADE': 'blue', 'SPADE': 'green'}
    
    plt.figure(figsize=(16, 8))
    
    # Estilos de línea para diferentes escenarios
    styles = {'Small': '-', 'Medium': '--', 'Full': '-.'}
    
    plotted_something = False
    
    for platform in dataframes:
        for scenario in dataframes[platform]:
            df = dataframes[platform][scenario]
            if df is not None:
                pid = pids[platform][scenario]
                cpu_cols = find_columns_by_pid(df, pid, '% Processor Time')
                
                # Check if CPU columns were found
                if not cpu_cols:
                    print(f"Warning: No CPU columns found for {platform} - {scenario}")
                    continue
                    
                # Check if ElapsedTime column exists
                if 'ElapsedTime' not in df.columns:
                    print(f"Warning: ElapsedTime column not found for {platform} - {scenario}")
                    continue
                
                for col in cpu_cols[:1]:  # Tomar solo la primera columna para no saturar la gráfica
                    # Verify the column actually exists
                    if col not in df.columns:
                        print(f"Warning: Column {col} not found in DataFrame for {platform} - {scenario}")
                        continue
                        
                    label = f"{platform} - {scenario}"
                    try:
                        plt.plot(df['ElapsedTime'], df[col], 
                                label=label, 
                                color=platform_colors[platform], 
                                linestyle=styles[scenario],
                                linewidth=2)
                        plotted_something = True
                    except Exception as e:
                        print(f"Error al graficar {platform} - {scenario}: {e}")
    
    if not plotted_something:
        print("Warning: No se pudo graficar ningún dato de CPU.")
        return
    
    plt.title('Comparación de Uso de CPU (Valores Brutos): JADE vs SPADE', fontsize=18)
    plt.xlabel('Tiempo Transcurrido (segundos)', fontsize=14)
    plt.ylabel('% Tiempo de Procesador (Bruto)', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=12)
    plt.tight_layout()
    
    try:
        # Guardar la gráfica
        plt.savefig('jade_vs_spade_cpu_comparison_raw.png', dpi=300)
        plt.show()
    except Exception as e:
        print(f"Error al guardar la gráfica: {e}")

# 🔧 FUNCIÓN CORREGIDA: Gráfica de CPU REAL (estilo Task Manager)
def plot_cpu_comparison_real(dataframes, platform_colors=None):
    """Gráfica de CPU con valores REALES (estilo Task Manager) - CORREGIDA"""
    if platform_colors is None:
        platform_colors = {'JADE': 'blue', 'SPADE': 'green'}
    
    plt.figure(figsize=(16, 8))
    styles = {'Small': '-', 'Medium': '--', 'Full': '-.'}
    plotted_something = False
    
    for platform in dataframes:
        for scenario in dataframes[platform]:
            df = dataframes[platform][scenario]
            if df is not None and 'ElapsedTime' in df.columns:
                pid = pids[platform][scenario]
                
                # 🔧 Buscar columnas de CPU REAL
                real_cpu_cols = find_real_cpu_columns(df, pid)
                
                if real_cpu_cols:
                    # Usar la primera columna de CPU real encontrada
                    col = real_cpu_cols[0]
                    label = f"{platform} - {scenario}"
                    try:
                        plt.plot(df['ElapsedTime'], df[col],
                                label=label, 
                                color=platform_colors[platform], 
                                linestyle=styles[scenario],
                                linewidth=2)
                        plotted_something = True
                        
                        # Debug info
                        avg_cpu = df[col].mean()
                        max_cpu = df[col].max()
                        print(f"✅ {label}: CPU Real Promedio = {avg_cpu:.2f}%, Máximo = {max_cpu:.2f}%")
                        
                    except Exception as e:
                        print(f"❌ Error graficando {platform} - {scenario}: {e}")
                else:
                    print(f"⚠️  No se encontraron columnas de CPU real para {platform} - {scenario}")
    
    if not plotted_something:
        print("❌ No se pudo graficar ningún dato de CPU real")
        return
    
    plt.title(f'Comparación de Uso de CPU REAL: JADE vs SPADE\n({LOGICAL_PROCESSORS} procesadores lógicos)', fontsize=18)
    plt.xlabel('Tiempo Transcurrido (segundos)', fontsize=14)
    plt.ylabel('% CPU Real del Sistema', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=12)
    
    plt.ylim(0, 10)
    
    # Líneas de referencia más realistas para CPU real
    plt.axhline(y=6.25, color='lightgreen', linestyle=':', alpha=0.7, label='6.25% (1 núcleo)')
    plt.axhline(y=12.5, color='orange', linestyle=':', alpha=0.7, label='12.5% (2 núcleos)')
    plt.axhline(y=25, color='red', linestyle=':', alpha=0.7, label='25% (4 núcleos)')
    plt.axhline(y=50, color='darkred', linestyle=':', alpha=0.7, label='50% sistema')
    
    plt.tight_layout()
    
    try:
        plt.savefig('jade_vs_spade_cpu_REAL_comparison.png', dpi=300)
        plt.show()
        print("✅ Gráfica de CPU real guardada como 'jade_vs_spade_cpu_REAL_comparison.png'")
    except Exception as e:
        print(f"❌ Error al guardar: {e}")

# 🔧 FUNCIÓN NUEVA: Comparación lado a lado (Real vs Simple)
def plot_cpu_comparison_side_by_side(dataframes, platform_colors=None):
    """Comparación lado a lado: CPU Real vs Normalización Simple"""
    if platform_colors is None:
        platform_colors = {'JADE': 'blue', 'SPADE': 'green'}
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))
    styles = {'Small': '-', 'Medium': '--', 'Full': '-.'}
    
    # Subplot 1: CPU Real
    plotted_real = False
    for platform in dataframes:
        for scenario in dataframes[platform]:
            df = dataframes[platform][scenario]
            if df is not None and 'ElapsedTime' in df.columns:
                pid = pids[platform][scenario]
                real_cpu_cols = find_real_cpu_columns(df, pid)
                
                if real_cpu_cols:
                    col = real_cpu_cols[0]
                    label = f"{platform} - {scenario}"
                    ax1.plot(df['ElapsedTime'], df[col],
                            label=label, 
                            color=platform_colors[platform], 
                            linestyle=styles[scenario],
                            linewidth=2)
                    plotted_real = True
    
    if plotted_real:
        ax1.set_title('CPU Real (Task Manager Style)', fontsize=14)
        ax1.set_xlabel('Tiempo (segundos)', fontsize=12)
        ax1.set_ylabel('% CPU Real', fontsize=12)
        ax1.grid(True, alpha=0.3)
        ax1.legend(fontsize=10)
        ax1.axhline(y=25, color='red', linestyle=':', alpha=0.5)
    
    # Subplot 2: Normalización Simple
    plotted_simple = False
    for platform in dataframes:
        for scenario in dataframes[platform]:
            df = dataframes[platform][scenario]
            if df is not None and 'ElapsedTime' in df.columns:
                pid = pids[platform][scenario]
                cpu_cols = find_columns_by_pid(df, pid, '% Processor Time')
                
                if cpu_cols:
                    base_col = cpu_cols[0]
                    simple_col = base_col + '_Normalized_Simple'
                    
                    if simple_col in df.columns:
                        label = f"{platform} - {scenario}"
                        ax2.plot(df['ElapsedTime'], df[simple_col],
                                label=label, 
                                color=platform_colors[platform], 
                                linestyle=styles[scenario],
                                linewidth=2)
                        plotted_simple = True
    
    if plotted_simple:
        ax2.set_title('CPU Normalizado Simple (INCORRECTO)', fontsize=14)
        ax2.set_xlabel('Tiempo (segundos)', fontsize=12)
        ax2.set_ylabel('% CPU Normalizado Simple', fontsize=12)
        ax2.grid(True, alpha=0.3)
        ax2.legend(fontsize=10)
        ax2.axhline(y=25, color='red', linestyle=':', alpha=0.5)
    
    plt.suptitle('Comparación: Método Correcto vs Incorrecto', fontsize=16)
    plt.tight_layout()
    
    try:
        plt.savefig('cpu_comparison_real_vs_simple.png', dpi=300)
        plt.show()
        print("✅ Comparación guardada como 'cpu_comparison_real_vs_simple.png'")
    except Exception as e:
        print(f"❌ Error: {e}")

# Función para crear gráficas comparativas de memoria (sin cambios)
def plot_memory_comparison(dataframes, memory_metric='Working Set', platform_colors=None):
    """
    Crea una gráfica comparativa de uso de memoria entre JADE y SPADE para los tres escenarios
    
    Args:
        dataframes: diccionario de DataFrames organizados por plataforma y escenario
        memory_metric: métrica de memoria a comparar ('Working Set', 'Private Bytes', etc.)
        platform_colors: diccionario opcional de colores para cada plataforma
    """
    if platform_colors is None:
        platform_colors = {'JADE': 'blue', 'SPADE': 'green'}
    
    plt.figure(figsize=(16, 8))
    
    # Estilos de línea para diferentes escenarios
    styles = {'Small': '-', 'Medium': '--', 'Full': '-.'}
    
    plotted_something = False
    
    for platform in dataframes:
        for scenario in dataframes[platform]:
            df = dataframes[platform][scenario]
            if df is not None:
                pid = pids[platform][scenario]
                memory_cols = find_columns_by_pid(df, pid, memory_metric)
                
                # Check if memory columns were found
                if not memory_cols:
                    print(f"Warning: No memory columns found for {platform} - {scenario}")
                    continue
                    
                # Check if ElapsedTime column exists
                if 'ElapsedTime' not in df.columns:
                    print(f"Warning: ElapsedTime column not found for {platform} - {scenario}")
                    continue
                
                for col in memory_cols[:1]:  # Tomar solo la primera columna para no saturar la gráfica
                    # Verify the column actually exists
                    if col not in df.columns:
                        print(f"Warning: Column {col} not found in DataFrame for {platform} - {scenario}")
                        continue
                        
                    # Convertir bytes a MB para mejor legibilidad
                    label = f"{platform} - {scenario}"
                    try:
                        plt.plot(df['ElapsedTime'], df[col]/1024/1024, 
                                label=label, 
                                color=platform_colors[platform], 
                                linestyle=styles[scenario],
                                linewidth=2)
                        plotted_something = True
                    except Exception as e:
                        print(f"Error al graficar {platform} - {scenario}: {e}")
    
    if not plotted_something:
        print(f"Warning: No se pudo graficar ningún dato de memoria ({memory_metric}).")
        return
    
    plt.title(f'Comparación de Uso de Memoria ({memory_metric}): JADE vs SPADE', fontsize=18)
    plt.xlabel('Tiempo Transcurrido (segundos)', fontsize=14)
    plt.ylabel('Memoria (MB)', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.legend(fontsize=12)
    plt.tight_layout()
    
    try:
        # Guardar la gráfica
        plt.savefig(f'jade_vs_spade_{memory_metric.lower().replace(" ", "_")}_comparison.png', dpi=300)
        plt.show()
    except Exception as e:
        print(f"Error al guardar la gráfica: {e}")

# 🔧 FUNCIÓN ACTUALIZADA: Métricas con CPU Real
def calculate_performance_metrics_with_real_cpu(dataframes):
    """
    Calcula métricas de rendimiento incluyendo CPU Real
    
    Args:
        dataframes: diccionario de DataFrames organizados por plataforma y escenario
    
    Returns:
        Un DataFrame con métricas de rendimiento
    """
    metrics = []
    
    for platform in dataframes:
        for scenario in dataframes[platform]:
            df = dataframes[platform][scenario]
            if df is not None:
                try:
                    pid = pids[platform][scenario]
                    
                    # Obtener métricas de CPU (original/bruto)
                    cpu_cols = find_columns_by_pid(df, pid, '% Processor Time')
                    if cpu_cols and cpu_cols[0] in df.columns:
                        cpu_col = cpu_cols[0]  # Usar la primera columna de CPU
                        cpu_avg_raw = df[cpu_col].mean()
                        cpu_max_raw = df[cpu_col].max()
                        cpu_std_raw = df[cpu_col].std()
                    else:
                        cpu_avg_raw = cpu_max_raw = cpu_std_raw = float('nan')
                    
                    # 🔧 Obtener métricas de CPU REAL
                    real_cpu_cols = find_real_cpu_columns(df, pid)
                    if real_cpu_cols and real_cpu_cols[0] in df.columns:
                        real_cpu_col = real_cpu_cols[0]
                        cpu_avg_real = df[real_cpu_col].mean()
                        cpu_max_real = df[real_cpu_col].max()
                        cpu_std_real = df[real_cpu_col].std()
                    else:
                        cpu_avg_real = cpu_max_real = cpu_std_real = float('nan')
                    
                    # Normalización simple (para comparación)
                    cpu_avg_simple = cpu_avg_raw / LOGICAL_PROCESSORS
                    cpu_max_simple = cpu_max_raw / LOGICAL_PROCESSORS
                    
                    # Obtener métricas de memoria privada
                    mem_private_cols = find_columns_by_pid(df, pid, 'Private Bytes')
                    if mem_private_cols and mem_private_cols[0] in df.columns:
                        mem_private_col = mem_private_cols[0]
                        mem_private_avg = df[mem_private_col].mean() / (1024 * 1024)  # En MB
                        mem_private_max = df[mem_private_col].max() / (1024 * 1024)  # En MB
                    else:
                        mem_private_avg = mem_private_max = float('nan')
                    
                    # Obtener métricas de memoria de trabajo
                    mem_ws_cols = find_columns_by_pid(df, pid, 'Working Set')
                    if mem_ws_cols and mem_ws_cols[0] in df.columns:
                        mem_ws_col = mem_ws_cols[0]
                        mem_ws_avg = df[mem_ws_col].mean() / (1024 * 1024)  # En MB
                        mem_ws_max = df[mem_ws_col].max() / (1024 * 1024)  # En MB
                    else:
                        mem_ws_avg = mem_ws_max = float('nan')
                    
                    # Calcular duración total
                    if 'ElapsedTime' in df.columns:
                        duration = df['ElapsedTime'].max() - df['ElapsedTime'].min()
                    else:
                        duration = float('nan')
                    
                    # 🔧 Agregar métricas con CPU REAL
                    metrics.append({
                        'Platform': platform,
                        'Scenario': scenario,
                        'CPU_Raw_Avg (%)': cpu_avg_raw,
                        'CPU_Raw_Max (%)': cpu_max_raw,
                        'CPU_Raw_StdDev': cpu_std_raw,
                        'CPU_Real_Avg (%)': cpu_avg_real,           # ✅ NUEVO: CPU Real
                        'CPU_Real_Max (%)': cpu_max_real,           # ✅ NUEVO: CPU Real 
                        'CPU_Real_StdDev': cpu_std_real,            # ✅ NUEVO: CPU Real
                        'CPU_Simple_Avg (%)': cpu_avg_simple,       # Para comparación
                        'CPU_Simple_Max (%)': cpu_max_simple,       # Para comparación
                        'Mem_Private_Avg (MB)': mem_private_avg,
                        'Mem_Private_Max (MB)': mem_private_max,
                        'Mem_WorkingSet_Avg (MB)': mem_ws_avg,
                        'Mem_WorkingSet_Max (MB)': mem_ws_max,
                        'Duration (sec)': duration
                    })
                except Exception as e:
                    print(f"Error al calcular métricas para {platform} - {scenario}: {e}")
    
    return pd.DataFrame(metrics) if metrics else pd.DataFrame()

# 🔧 FUNCIÓN ACTUALIZADA: Análisis estadístico con CPU Real
def perform_statistical_analysis_with_real_cpu(metrics_df):
    """
    Realiza un análisis estadístico comparativo entre JADE y SPADE usando CPU Real
    
    Args:
        metrics_df: DataFrame con métricas calculadas incluyendo CPU Real
    """
    if metrics_df.empty:
        print("No hay suficientes datos para realizar un análisis estadístico.")
        return
    
    try:
        # Análisis por plataforma
        if 'Platform' in metrics_df.columns:
            platform_comparison = metrics_df.groupby('Platform').mean()
            if 'Duration (sec)' in platform_comparison.columns:
                platform_comparison = platform_comparison.drop('Duration (sec)', axis=1)  # Eliminar duración para evitar contaminar el análisis
            
            print("\n=== Comparación de Rendimiento Promedio por Plataforma (CON CPU REAL) ===")
            print(platform_comparison)
        else:
            print("Error: No se encontró la columna 'Platform' en el DataFrame de métricas.")
        
        # Análisis por escenario y plataforma
        if 'Scenario' in metrics_df.columns and 'Platform' in metrics_df.columns:
            # Verificar que existan las columnas necesarias (incluyendo CPU Real)
            required_cols = ['CPU_Real_Avg (%)', 'CPU_Real_Max (%)', 'CPU_Simple_Avg (%)', 'Mem_WorkingSet_Avg (MB)', 'Duration (sec)']
            missing_cols = [col for col in required_cols if col not in metrics_df.columns]
            
            if missing_cols:
                print(f"Advertencia: Faltan las siguientes columnas para el análisis: {missing_cols}")
                available_cols = [col for col in required_cols if col in metrics_df.columns]
                if not available_cols:
                    print("No hay columnas disponibles para el análisis por escenario y plataforma.")
                    return
                
                scenario_platform_comparison = metrics_df.pivot_table(
                    index='Scenario', 
                    columns='Platform',
                    values=available_cols
                )
            else:
                scenario_platform_comparison = metrics_df.pivot_table(
                    index='Scenario', 
                    columns='Platform',
                    values=required_cols
                )
            
            print("\n=== Comparación de Rendimiento por Escenario y Plataforma (CON CPU REAL) ===")
            print(scenario_platform_comparison)
        else:
            print("Error: No se encontraron las columnas 'Scenario' o 'Platform' en el DataFrame de métricas.")
        
        # 🔧 Calcular mejora relativa usando CPU REAL
        if 'JADE' in metrics_df['Platform'].values and 'SPADE' in metrics_df['Platform'].values:
            relative_improvement = pd.DataFrame()
            
            for scenario in metrics_df['Scenario'].unique():
                jade_data = metrics_df[(metrics_df['Platform'] == 'JADE') & (metrics_df['Scenario'] == scenario)]
                spade_data = metrics_df[(metrics_df['Platform'] == 'SPADE') & (metrics_df['Scenario'] == scenario)]
                
                if not jade_data.empty and not spade_data.empty:
                    jade_row = jade_data.iloc[0]
                    spade_row = spade_data.iloc[0]
                    
                    # Crear diccionario para almacenar las métricas
                    improvement_data = {'Scenario': scenario}
                    
                    # 🔧 Para CPU REAL, menor es mejor
                    if 'CPU_Real_Avg (%)' in jade_row and 'CPU_Real_Avg (%)' in spade_row and pd.notna(spade_row['CPU_Real_Avg (%)']) and spade_row['CPU_Real_Avg (%)'] != 0:
                        cpu_real_diff = (spade_row['CPU_Real_Avg (%)'] - jade_row['CPU_Real_Avg (%)']) / spade_row['CPU_Real_Avg (%)'] * 100
                        improvement_data['CPU Real Improvement (%)'] = cpu_real_diff
                    
                    # Para CPU simple (para comparación)
                    if 'CPU_Simple_Avg (%)' in jade_row and 'CPU_Simple_Avg (%)' in spade_row and pd.notna(spade_row['CPU_Simple_Avg (%)']) and spade_row['CPU_Simple_Avg (%)'] != 0:
                        cpu_simple_diff = (spade_row['CPU_Simple_Avg (%)'] - jade_row['CPU_Simple_Avg (%)']) / spade_row['CPU_Simple_Avg (%)'] * 100
                        improvement_data['CPU Simple Improvement (%)'] = cpu_simple_diff
                    
                    # Para memoria
                    if 'Mem_WorkingSet_Avg (MB)' in jade_row and 'Mem_WorkingSet_Avg (MB)' in spade_row and pd.notna(spade_row['Mem_WorkingSet_Avg (MB)']) and spade_row['Mem_WorkingSet_Avg (MB)'] != 0:
                        mem_diff = (spade_row['Mem_WorkingSet_Avg (MB)'] - jade_row['Mem_WorkingSet_Avg (MB)']) / spade_row['Mem_WorkingSet_Avg (MB)'] * 100
                        improvement_data['Memory Improvement (%)'] = mem_diff
                    
                    # Para duración, menor es mejor
                    if 'Duration (sec)' in jade_row and 'Duration (sec)' in spade_row and pd.notna(spade_row['Duration (sec)']) and spade_row['Duration (sec)'] != 0:
                        time_diff = (spade_row['Duration (sec)'] - jade_row['Duration (sec)']) / spade_row['Duration (sec)'] * 100
                        improvement_data['Time Improvement (%)'] = time_diff
                    
                    # Solo agregar si hay al menos una métrica de mejora
                    if len(improvement_data) > 1:  # Más que solo 'Scenario'
                        relative_improvement = pd.concat([relative_improvement, pd.DataFrame([improvement_data])], ignore_index=True)
            
            if not relative_improvement.empty:
                print("\n=== 🔧 Mejora Relativa de JADE vs SPADE usando CPU REAL (% de mejora) ===")
                print("Nota: Valores positivos = JADE mejor, valores negativos = SPADE mejor")
                print(relative_improvement)
                
                # 🔧 Crear visualización con CPU REAL
                if len(relative_improvement) > 0:
                    plt.figure(figsize=(14, 8))
                    
                    scenarios = relative_improvement['Scenario']
                    
                    # Determinar qué columnas están disponibles
                    improvement_cols = []
                    if 'CPU Real Improvement (%)' in relative_improvement.columns:
                        improvement_cols.append(('CPU Real Improvement (%)', 'CPU Real', '#3498db'))
                    if 'CPU Simple Improvement (%)' in relative_improvement.columns:
                        improvement_cols.append(('CPU Simple Improvement (%)', 'CPU Simple', '#95a5a6'))
                    if 'Memory Improvement (%)' in relative_improvement.columns:
                        improvement_cols.append(('Memory Improvement (%)', 'Memoria', '#2ecc71'))
                    if 'Time Improvement (%)' in relative_improvement.columns:
                        improvement_cols.append(('Time Improvement (%)', 'Tiempo', '#e74c3c'))
                    
                    if not improvement_cols:
                        print("No hay columnas de mejora disponibles para visualizar.")
                        return
                    
                    x = np.arange(len(scenarios))
                    width = 0.8 / len(improvement_cols)
                    
                    # Crear barras
                    for i, (col, label, color) in enumerate(improvement_cols):
                        values = relative_improvement[col]
                        pos = x + width * (i - len(improvement_cols) / 2 + 0.5)
                        bars = plt.bar(pos, values, width, label=label, color=color)
                        
                        # Añadir etiquetas de valor a las barras
                        for j, v in enumerate(values):
                            if pd.notna(v):
                                plt.text(pos[j], v + (5 if v >= 0 else -10), f"{v:.1f}%", ha='center', fontsize=9)
                    
                    # Añadir detalles a la gráfica
                    plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
                    plt.xlabel('Escenario', fontsize=14)
                    plt.ylabel('Mejora (%) - JADE vs SPADE', fontsize=14)
                    plt.title('🔧 Comparación de Rendimiento: JADE vs SPADE\n(Usando CPU REAL - Task Manager Style)', fontsize=16)
                    plt.xticks(x, scenarios)
                    plt.legend()
                    
                    plt.tight_layout()
                    
                    try:
                        plt.savefig('jade_vs_spade_improvement_REAL_CPU.png', dpi=300)
                        plt.show()
                        print("✅ Gráfica de mejora con CPU real guardada")
                    except Exception as e:
                        print(f"❌ Error al guardar la gráfica de mejora: {e}")
            else:
                print("No hay suficientes datos para calcular la mejora relativa entre plataformas.")
        else:
            print("No hay datos para ambas plataformas (JADE y SPADE) para realizar una comparación.")
    
    except Exception as e:
        print(f"Error durante el análisis estadístico: {e}")

# 🔧 FUNCIÓN PRINCIPAL ACTUALIZADA
def main():
    print("=== 🔧 Análisis Comparativo JADE vs SPADE con CPU REAL ===")
    print(f"Sistema: {LOGICAL_PROCESSORS} procesadores lógicos")
    print("NUEVA: Implementación de CPU real estilo Task Manager")
    
    # Cargar los datos
    loaded_data = {}
    for platform in data_files:
        loaded_data[platform] = {}
        for scenario in data_files[platform]:
            file_path = data_files[platform][scenario]
            loaded_data[platform][scenario] = load_perfmon_data(file_path)
    
    # Verificar si se cargaron datos correctamente
    data_loaded = False
    for platform in loaded_data:
        for scenario in loaded_data[platform]:
            if loaded_data[platform][scenario] is not None:
                data_loaded = True
                break
        if data_loaded:
            break
    
    if not data_loaded:
        print("❌ Error! No se pudo cargar ningún archivo de datos. Verifica las rutas y formatos.")
        return
    
    # Colores para cada plataforma
    platform_colors = {'JADE': '#1e88e5', 'SPADE': '#43a047'}
    
    try:
        # 🔧 Generar visualizaciones comparativas CON CPU REAL
        print("\n=== 🔧 Generando Visualizaciones con CPU REAL ===")
        
        # Gráfica original de CPU (valores brutos)
        print("1. Gráfica CPU original (valores brutos)...")
        plot_cpu_comparison(loaded_data, platform_colors)
        
        # 🔧 NUEVA: Gráfica de CPU REAL (estilo Task Manager)
        print("2. Gráfica CPU REAL (Task Manager)...")
        plot_cpu_comparison_real(loaded_data, platform_colors)
        
        # 🔧 NUEVA: Comparación lado a lado
        print("3. Comparación lado a lado (Real vs Simple)...")
        plot_cpu_comparison_side_by_side(loaded_data, platform_colors)
        
        # Gráficas de memoria (sin cambios)
        print("4. Gráficas de memoria...")
        plot_memory_comparison(loaded_data, 'Working Set', platform_colors)
        plot_memory_comparison(loaded_data, 'Private Bytes', platform_colors)
        
        # 🔧 Calcular métricas de rendimiento CON CPU REAL
        print("\n=== 🔧 Calculando Métricas con CPU REAL ===")
        metrics_df = calculate_performance_metrics_with_real_cpu(loaded_data)
        
        if not metrics_df.empty:
            print("\n✅ Resumen de Métricas de Rendimiento (CON CPU REAL):")
            print(metrics_df)
            
            # Guardar resultados en CSV
            try:
                metrics_df.to_csv('jade_vs_spade_metrics_REAL_CPU.csv', index=False)
                print("✅ Métricas guardadas en 'jade_vs_spade_metrics_REAL_CPU.csv'")
            except Exception as e:
                print(f"❌ Error al guardar las métricas en CSV: {e}")
            
            # 🔧 Realizar análisis estadístico CON CPU REAL
            print("\n=== 🔧 Análisis Estadístico con CPU REAL ===")
            perform_statistical_analysis_with_real_cpu(metrics_df)
        else:
            print("❌ No se pudieron calcular métricas de rendimiento debido a problemas con los datos.")
        
        print("\n=== ✅ Análisis Completado con CPU REAL ===")
        print("📊 Visualizaciones generadas:")
        print("   - jade_vs_spade_cpu_comparison_raw.png (valores brutos)")
        print("   - jade_vs_spade_cpu_REAL_comparison.png (CPU real - Task Manager)")
        print("   - cpu_comparison_real_vs_simple.png (comparación métodos)")
        print("   - jade_vs_spade_improvement_REAL_CPU.png (mejoras con CPU real)")
        print("📈 Métricas guardadas en: jade_vs_spade_metrics_REAL_CPU.csv")
        print(f"🔧 CPU real calculada usando fórmula: (proceso_cpu / ({LOGICAL_PROCESSORS} × total_cpu)) × 100")
        
    except Exception as e:
        print(f"❌ Error durante la ejecución del análisis: {e}")
        print("Parte del análisis puede haberse completado, por favor revise los archivos generados.")

# 🔧 FUNCIÓN ADICIONAL: Verificar datos cargados
def debug_loaded_data(loaded_data):
    """Función de debug para verificar qué columnas se cargaron"""
    print("\n=== 🔍 DEBUG: Verificando datos cargados ===")
    
    for platform in loaded_data:
        for scenario in loaded_data[platform]:
            df = loaded_data[platform][scenario]
            if df is not None:
                print(f"\n📂 {platform} - {scenario}:")
                print(f"   Filas: {len(df)}")
                
                # Buscar columnas relevantes
                cpu_cols = [col for col in df.columns if '% Processor Time' in col or 'tiempo de procesador' in col.lower()]
                total_cols = [col for col in cpu_cols if '_Total' in col]
                real_cols = [col for col in df.columns if '_Real' in col]
                
                print(f"   Columnas CPU: {len(cpu_cols)}")
                print(f"   Columnas _Total: {len(total_cols)}")
                print(f"   Columnas _Real: {len(real_cols)}")
                
                if total_cols:
                    print(f"   ✅ _Total encontrado: {total_cols[0]}")
                else:
                    print(f"   ❌ No se encontró _Total")
                
                if real_cols:
                    print(f"   ✅ CPU Real creado: {real_cols[:2]}")  # Mostrar primeras 2
                else:
                    print(f"   ❌ No se creó CPU Real")
            else:
                print(f"\n❌ {platform} - {scenario}: No se pudo cargar")

if __name__ == "__main__":
    main()
    
    # Uncomment para debug
    # debug_loaded_data(loaded_data)
```
